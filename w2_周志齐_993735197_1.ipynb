    正常值: 空腹血糖: 60~110 mg/dl; 餐后2h血糖 <140 mg/dl
    糖尿病: 空腹血糖: >=126 mg/dl; 餐后2h血糖 >=200 mg/dl
    0: 0-90
    1: 90-140
    2: 140-200
    3: >200

3.BloodPressure: 舒张压（mm Hg）

    低压, (收缩压为高压)
    高大,肥胖者血压较高,  男女不同, 与妊娠者分开
    正常值: 60-89, 高血压:>=90, 低血压:<60
    0: <60
    1: 60-89
    2: >90
    不可能出现0

4.SkinThickness: 皮褶厚度（mm）

    可以推断全身脂肪含量
    正常: <40,(中国)
    男女不同

5.Insulin: 2小时血清胰岛素（mu U / ml）

    糖尿病是由于胰腺不能产生足够的胰岛素
    或者身体的细胞对所产生的胰岛素没有适当的反应
    空腹者一般应介于5pmol/L到20pmol/L,餐后1小时,为空腹5～6倍；餐后2小时为空腹4～5倍；餐后3小时为空腹2.5～3倍.
    餐后2小时胰岛素正常值是不高于11.1mmol/l 大概不高于77.3mu U / ml

6.BMI: 体重指数（以千克/身高（m）为单位）^ 2）

    正常: <24 kg/m2
    超重: 24-28 kg/m2
    肥胖: >28 kg/m2

7.DiabetesPedigreeFunction: 糖尿病谱系功能

    糖尿病家庭病史
    看不懂, 以0.5分类吧
    0: <0.5
    1: 0.5-1.0
    2: 1.0-1.5
    3: >1.5

8.Age: 年龄（岁）

    1: 20-30
    2: 30-40
    3: 40-60
    4: 60-

9.Outcome: 类变量（0或1）

    1型: 大多数是在儿童中, 20岁以下
    2型: 最常见的糖尿病类型, 大多在40岁以上, 生活方式因素和遗传因素, 肥胖,缺乏体力活动,饮食不当,应力,和城市化, 
    妊娠糖尿病: 发生在所有怀孕的约2-10％，并可能在分娩后改善或消失 
    其他特定类型: 常染色体显性遗传形式, 前驱糖尿病, 成人隐匿性自身免疫性糖尿病成人隐匿性自身免疫性糖尿病

## 调优

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

#需要调优的参数
# 请尝试将L1正则和L2正则分开，并配合合适的优化求解算法（slover）
#tuned_parameters = {'penalty':['l1','l2'],
#                   'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000]
#                   }
penaltys = ['l1','l2']
Cs = [0.001, 0.01, 0.1, 1, 10, 100, 1000]
tuned_parameters = dict(penalty = penaltys, C = Cs)

lr_penalty= LogisticRegression()
grid= GridSearchCV(lr_penalty, tuned_parameters,cv=5, scoring='neg_log_loss')
grid.fit(X_train,y_train)
```

```python
grid.cv_results_
```

```python
def fit_grid_point_Linear(C, X_train, y_train, X_val, y_val):
    
    # 在训练集是那个利用SVC训练
    SVC2 =  LinearSVC( C = C)
    SVC2 = SVC2.fit(X_train, y_train)
    
    # 在校验集上返回accuracy
    accuracy = SVC2.score(X_val, y_val)
    
    print("accuracy: {}".format(accuracy))
    return accuracy
```

```python
def fit_grid_point_RBF(C, gamma, X_train, y_train, X_val, y_val):
    
    # 在训练集是那个利用SVC训练
    SVC3 =  SVC( C = C, kernel='rbf', gamma = gamma)
    SVC3 = SVC3.fit(X_train, y_train)
    
    # 在校验集上返回accuracy
    accuracy = SVC3.score(X_val, y_val)
    
    print("accuracy: {}".format(accuracy))
    return accuracy
```

```python
#需要调优的参数
C_s = np.logspace(-2, 2, 5)# logspace(a,b,N)把10的a次方到10的b次方区间分成N份 
gamma_s = np.logspace(-2, 2, 5)  

accuracy_s = []
for i, oneC in enumerate(C_s):
    for j, gamma in enumerate(gamma_s):
        tmp = fit_grid_point_RBF(oneC, gamma, X_train, y_train, X_val, y_val)
        accuracy_s.append(tmp)
```

```python
accuracy_s1 =np.array(accuracy_s).reshape(len(C_s),len(gamma_s))
x_axis = np.log10(C_s)
for j, gamma in enumerate(gamma_s):
    plt.plot(x_axis, np.array(accuracy_s1[:,j]), label = ' Test - log(gamma)' + str(np.log10(gamma)))

plt.legend()
plt.xlabel( 'log(C)' )                                                                                                      
plt.ylabel( 'accuracy' )
plt.savefig('RBF_SVM_Otto.png' )

plt.show()
```
